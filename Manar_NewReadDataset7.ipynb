```python
import numpy as np
from numpy import mean
from numpy import std
from numpy import dstack
from pandas import read_csv
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import LSTM
from tensorflow.keras.utils import to_categorical
#from matplotlib import pyplot
from numpy import save
```

```python
def data_normalize(data_temp):
    data_temp2=data_temp.T
    data_temp2 -=np.mean(data_temp2,axis=0)
    data_temp2 /=np.std(data_temp2,axis=0)
    data_temp=data_temp2.T
    return data_temp
```

```python
def load_file(filepath):
	dataframe = read_csv(filepath, header=None, delim_whitespace=True)
	return dataframe.values
```

```python
def load_group(filenames, prefix=''):
	loaded = list()
	for name in filenames:
		data = load_file(prefix + name)
		loaded.append(data)
	# stack group so that features are the 3rd dimension
	loaded = dstack(loaded)
	return loaded
```

```python
def load_dataset_group(group, prefix=''):
    filepathData = prefix + group +'/data/'
    filepathLabel = prefix + group +'/label/'
    # load all 9 files as a single array
    filenames = list()
    # body acceleration
    fileNamesX = []
    fileNamesY = []
    #fileNamesTrain.append(12)
    if (group == "train"):
        for x in range(519):
            fileNamesX.append(str(x+1)+'.txt')
        for x in range(519):
            fileNamesY.append(str(x+1)+'_label.txt')
    if (group == "test"):
        for y in range(58):
            fileNamesX.append(str(y+1)+'.txt')
        for y in range(58):
            fileNamesY.append(str(y+1)+'_label.txt')

    # load input data
    X = load_group(fileNamesX, filepathData)
    # load class output
    y = load_group(fileNamesY, filepathLabel)
    return X, y
```

```python
def load_dataset(prefix=''):
	# load all train
	trainX, trainy = load_dataset_group('train', prefix + 'Dataset7/')
	print(trainX.shape, trainy.shape)
	# load all test
	testX, testy = load_dataset_group('test', prefix + 'Dataset7/')
	print(testX.shape, testy.shape)
	# zero-offset class values
	trainy = trainy - 1
	testy = testy - 1
	# one hot encode y
	trainy = to_categorical(trainy)
	testy = to_categorical(testy)
	print(trainX.shape, trainy.shape, testX.shape, testy.shape)
	return trainX, trainy, testX, testy
```

```python
trainX, trainy, testX, testy=load_dataset()
```
